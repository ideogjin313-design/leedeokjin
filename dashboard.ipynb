```python
import streamlit as st
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go

# 페이지 설정
st.set_page_config(page_title="QA/QC 품질 관리 대시보드", layout="wide")

# 타이틀 부분
st.title("🧪 QA/QC 품질 관리 대시보드")
st.caption("Merged 데이터 기반 - Itchy Scalp & Cool Menthol 제품군")

# --- 상단 지표 영역 (Row 1) ---
col1, col2, col3 = st.columns([1, 1, 2])

with col1:
    st.info("현재 진행 배치 ID")
    st.subheader("Batch #8")

with col2:
    st.info("🌡️ 온도 상태 모니터링")
    st.metric(label="현재 온도", value="26.5 °C", delta="적정", delta_color="normal")
    st.progress(0.65) # 예시 바

with col3:
    st.success("✅ 안정성 테스트 현황 (Stability Test)")
    c1, c2 = st.columns(2)
    c1.metric("합격", "30건", "93.75%")
    c2.metric("불합격", "2건", "-6.25%", delta_color="inverse")

st.divider()

# --- 중간 분석 영역 (Row 2) ---
col4, col5 = st.columns([1, 1])

with col4:
    st.subheader("📦 품질 합격 현황")
    # 점도 상태 분포 (막대 그래프)
    viscosity_data = pd.DataFrame({
        "상태": ["묽음(Low)", "안정(Target)", "떡짐(High)"],
        "비율": [15, 70, 15]
    })
    fig_vis = px.bar(viscosity_data, x="비율", y="상태", orientation='h',
                     color="상태", color_discrete_map={"묽음(Low)": "#636EFA", "안정(Target)": "#00CC96", "떡짐(High)": "#EF553B"})
    st.plotly_chart(fig_vis, use_container_width=True)

with col5:
    st.subheader("📈 레올로지 골든 배치 비교")
    # 골든 배치 타겟 밴드와 현재 배치 곡선 시뮬레이션
    fig_rheo = go.Figure()
    # 타겟 영역 (Golden Batch Band)
    fig_rheo.add_trace(go.Scatter(x=[1, 2, 3, 4, 5], y=[10, 8, 6, 4, 2], fill='toself', name='Golden Band', line_color='rgba(255,255,0,0.2)'))
    # 현재 배치 곡선
    fig_rheo.add_trace(go.Scatter(x=[1, 2, 3, 4, 5], y=[9.5, 7.8, 5.8, 3.9, 1.8], mode='lines+markers+text', name='Current Batch', text=["8"]*5))
    fig_rheo.update_layout(xaxis_title="Log 전단속도", yaxis_title="Log 점도")
    st.plotly_chart(fig_rheo, use_container_width=True)

# --- 하단 성분 분석 영역 (Row 3) ---
st.divider()
st.subheader("🔍 비정상 점도 성분 및 민감도 분석")

col6, col7 = st.columns([2, 1])

with col6:
    # 누적 막대 그래프 데이터
    ingredients = ["Arlypon TT", "Plantacare 818", "Citric Acid", "Glycerin", "Pliater", "Phenoxyethanol"]
    df_comp = pd.DataFrame({
        "Type": ["Thinned-Out", "Thickened-Up"] * 2,
        "Ingredient": ingredients[0]*4, # 예시용 데이터 구성
        "Value": [10, 20, 30, 40] 
    })
    # 이미지의 하단 차트를 Plotly Stacked Bar로 구현
    st.write("합격 배치 중 비정상 점도 성분 분석")
    # (실제 구현 시에는 더 구체적인 DataFrame 작성이 필요합니다)
    st.image("https://via.placeholder.com/800x200.png?text=Stacked+Bar+Chart+for+Ingredients") # 대체 이미지 또는 Plotly 코드

with col7:
    st.write("⚠️ 성분 민감도 리모트")
    sensitivity_df = pd.DataFrame([
        {"성분": "Arlypon TT", "문제 기여도": "65%", "원인": "배합 편차로 인한 묽음"},
        {"성분": "Plantacare 818", "문제 기여도": "25%", "원인": "부정확한 중화"},
        {"성분": "Citric Acid", "문제 기여도": "25%", "원인": "중화로 인한 떡짐"}
    ])
    st.table(sensitivity_df)
```
